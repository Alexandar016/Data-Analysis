## Ex. 2 Bayesian probability

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# p is the probability of being infected, num_test is the number of positive tests
def infections(p, num_test): 
    return ((0.96**num_test)*p)/((0.96**num_test)*p+(0.05**num_test)*(1-p))
```

```python
plt.plot(np.arange(0,1,0.01), infections(np.arange(0,1,0.01), 1), label = 'one positive test')
plt.plot(np.arange(0,1,0.01), infections(np.arange(0,1,0.01), 2), label = 'two positive tests', linestyle = '--')
plt.xlabel('P(infected)')
plt.ylabel('P(infected|+)')
plt.legend()
plt.show()
```

```python
infections(0.001,1)
```

```python
infections(0.1,1)
```
