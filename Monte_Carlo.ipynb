```python
import numpy 
import matplotlib.pyplot as plt
```

```python
#definition of a chisquare distribution
def chi(N, mu, var, ndf):
    res = numpy.zeros(N)
    for m in range(0,N):
        summe = 0
        
        for i in range(0, ndf):
            summe += (numpy.random.normal(mu, var))**2
        
        res[m] = summe
        
    return res  
```

```python
#set chi_i, so results after this step are reproducable
#mu = 0, sigma = 1 for standard normal distribution
chi_2 = chi(10000, 0, 1, 2)
chi_3 = chi(10000, 0, 1, 3)
chi_5 = chi(10000, 0, 1, 5)
chi_10 =chi(10000, 0,1, 10)
```

```python
plt.hist(chi_2,color='blue', bins = 50, label = r'ndf =2, $\sigma^2$ = '+ str(round((numpy.std(chi_2))**2,2)) + r', $\mu$  = ' + str(round(numpy.mean(chi_2),2)))
plt.legend()
plt.show()
```

```python
plt.hist(chi_3, color='blue',bins = 50, label = r'ndf =3, $\sigma^2$ = '+ str(round((numpy.std(chi_3))**2,2)) + r', $\mu$  = ' + str(round(numpy.mean(chi_3),2)))
plt.legend()
plt.show()
```

```python
plt.hist(chi_5,color='blue', bins = 50, label = r'ndf =5, $\sigma^2$ = '+ str(round((numpy.std(chi_5))**2,2)) + r', $\mu$  = ' + str(round(numpy.mean(chi_5),2)))
plt.legend()
plt.show()
```

```python
plt.hist(chi_10,color='blue', bins = 50, label = r'ndf =10, $\sigma^2$ = '+ str(round((numpy.std(chi_10))**2,2)) + r', $\mu$  = ' + str(round(numpy.mean(chi_10),2)))
plt.legend()
plt.show()
```

The theoretical values :

$\mu$ = ndf
