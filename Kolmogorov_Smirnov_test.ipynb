```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
import scipy.stats as st
%matplotlib inline
```

```python
# Given values for N
N = [10, 100, 1000]
```

## a)

```python
# Create list "rand_arr", which stores all samples for all N-values
rand_arr = []

# The for-loop goes through the N-values
for i in np.arange(0,3):
    # The list "temp" will contain the 1000 samples (first dimension) and each entry of "temp"
    # contains the N elements
    temp = []
    # Fill each entry with the uniformly distributed random numbers 
    for k in np.arange(0, 1000):
        temp.append(np.random.uniform(size=N[i]))
    # Append "temp" to "rand_arr"
    rand_arr.append(temp)
```

## b)

```python
# Lists that store the distances and probabilities for all samples of all N-values 
d_arr = []
p_ks_arr = []

# Similar to the way rand_arr was filled
for i in np.arange(0,3):
    d_temp = []
    p_temp = []
    for k in np.arange(0,1000):
        # Use scipy.stats.kstest to get the distance and probability with respect to the uniform distribution
        res = st.kstest(rand_arr[i][k], "uniform")
        d_temp.append(res[0])
        p_temp.append(res[1])
        
    d_arr.append(d_temp)
    p_ks_arr.append(p_temp)
```

## c)

```python
plt.figure(figsize=(16,9), dpi=100, facecolor="white")
gs = gridspec.GridSpec(2, 1)

# Plot the distances for all N-values
ax = plt.subplot(gs[0,0])
plt.plot(np.arange(1,1001), d_arr[0], ls="", marker="o", markersize=4, c="b", label="N = "+str(N[0]))
plt.plot(np.arange(1,1001), d_arr[1], ls="", marker="o", markersize=4, c="r", label="N = "+str(N[1]))
plt.plot(np.arange(1,1001), d_arr[2], ls="", marker="o", markersize=4, c="orange", label="N = "+str(N[2]))
plt.ylabel("${D}_{n}$", fontsize=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc="upper right", fontsize=18, framealpha=1, shadow=1)

# Plot the probabilities for all N-values
ax = plt.subplot(gs[1,0])
plt.plot(np.arange(1,1001), p_ks_arr[0], ls="", marker="o", markersize=4, c="b", label="N = "+str(N[0]))
plt.plot(np.arange(1,1001), p_ks_arr[1], ls="", marker="o", markersize=4, c="r", label="N = "+str(N[1]))
plt.plot(np.arange(1,1001), p_ks_arr[2], ls="", marker="o", markersize=4, c="orange", label="N = "+str(N[2]))    
plt.ylabel("${P}_{ks}$", fontsize=20)
plt.xlabel("Sample", fontsize=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
```
