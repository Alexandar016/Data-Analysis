```python
import numpy as np
import math
import matplotlib.pyplot as plt

def mean_calc(sample):
    return sum(sample) / len(sample)


f = open('hist.txt', "r")
lines = f.readlines()
result = []
for x in lines:
    result.append(float(x))
mean = mean_calc(result)
print("Mean is: ", mean)


# median

result.sort()
num = len(result)
if num % 2 == 0:
    median1 = result[num//2]
    median2 = result[num//2 - 1]
    median = (median1 + median2)/2
else:
    median = result[num//2]
print("Median is: ", median)


# Mode Most probable value (for histogram binnig-dependent!)
plt.title('bins=10', color='#555500', size=20)
plt.hist(result, bins=10)
plt.figure()

plt.title('bins=5', color='#555500', size=20)
plt.hist(result, bins=5)
plt.figure()

plt.title('bins=50', color='#555500', size=20)
plt.hist(result, bins=50)
plt.show()
```
