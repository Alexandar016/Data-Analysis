```python
import numpy as np
import math 
import matplotlib.pyplot as plt
```

```python
def mean_calc(sample):
    return sum(sample) / len(sample)
```

```python
f=open('covariance.txt',"r")
lines=f.readlines()
result1=[]
result2=[]

for x in lines:
    result1.append(float(x.split(',')[0]))
mean_x=mean_calc(result1)
mean_x_sq=mean_x**2

for y in lines:
    result2.append(float(y.split(',')[1]))
mean_y=mean_calc(result2)
mean_y_sq=mean_y**2

squared_numbers1 = [number1 ** 2 for number1 in result1]
squared_numbers2 = [number2 ** 2 for number2 in result2]

A_sq=mean_calc(squared_numbers1)
B_sq=mean_calc(squared_numbers2)

products = []
for num1, num2 in zip(result1, result2):
    products.append(num1 * num2)
mean_z=mean_calc(products)

print('Mean for variable 1: ',mean_x)
print('Mean for variable 2: ',mean_y)

A=np.zeros((2,2))
A[0][0]=A_sq-mean_x_sq
A[1][1]=B_sq-mean_y_sq
A[0][1]=mean_z-mean_x*mean_y
A[1][0]=mean_z-mean_x*mean_y
print('The covariance matrix is: \n',A)

B=np.zeros((2,2))
B[0][0]=1
B[1][1]=1
B[0][1]=A[0][1]/(math.sqrt(A[0][0]*A[1][1]))
B[1][0]=A[0][1]/(math.sqrt(A[0][0]*A[1][1]))
print('the correlation coefficient matrix: \n',B)

#f.close()
```
