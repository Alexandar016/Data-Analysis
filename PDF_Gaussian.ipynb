```python
import math
import numpy as np
import scipy.integrate as integrate
import matplotlib
import matplotlib.pyplot as plt
# Disribution
def g(x):
    if (x>= 5 and x<= 12):
        return 1./7.
    else:
        return 0

def gaussian(x, mu, sigma):
    return np.exp(-(x - mu)**2 / (2 * sigma**2)) / np.sqrt((2*np.pi*sigma**2))

#MEAN AND VARIANCE based from paper calculation
mean_x=17/2
variance_x=49/12

#PLOTING THE STUFF
fig = plt.figure(figsize=(10,10))

plt.xlabel(r'$x$', size=17)
plt.ylabel(r'$g(x)$ ', size=17 )
x_values = np.linspace(-10, 20, 1000)
y=np.zeros(len(x_values))
ygauss=np.zeros(len(x_values))
#calc y values
for j in range(0,len(x_values)):
    y[j]=g(x_values[j])
    ygauss[j]=gaussian(x_values[j] , mean_x , variance_x)
    
plt.plot(x_values,y, marker='', linestyle='solid', color='blue', lw=2, label=r'$\mathrm{Distribution}$' )
plt.plot(x_values, ygauss,linestyle='solid', color='green', lw=2, label=r'$\mathrm{Gauss}$')
plt.errorbar(mean_x, 1./7., xerr=variance_x, yerr=0., marker='o', linestyle='', color='red', markersize=10.0,  capsize=10.0, elinewidth=4, label=r'$\mathrm{Mean \;and\; standard\; deviation}$')
plt.grid()
plt.legend(ncol=1, prop={"size":12}, loc=2, handlelength=2.5, numpoints=1)
plt.show()
```
