```python
import numpy as np
from scipy.stats import moment
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_arr = np.genfromtxt("moment_distributions.txt")
data_arr.shape
```

```python
def moments(data):
    results = np.empty(4)
    for i in np.arange(0,4):
        results[i] = moment(data, moment=i+1)
        
    return results
```

```python
moments_arr = np.empty((5,4))

for i in np.arange(0,5):
    moments_arr[i] = moments(data_arr[i])
    
moments_arr
#moments_arr.shape
```

```python
diffs = np.empty((4,4))

for i in np.arange(0,4):
    diffs[i] = moments_arr[i+1] - moments_arr[0]

diffs
```

```python
plt.rcParams['axes.labelpad']='20'
plt.rcParams['xtick.major.pad']='10'
plt.rcParams['ytick.major.pad']='10'
```

```python
plt.figure(figsize=(7, 9), dpi=80, facecolor="white")

for i in np.arange(0,5):
    plt.plot(np.arange(1,5), moments_arr[i], label="Distribution "+str(i+1), ls="", marker="o")

plt.xlabel("k", fontsize=15)
plt.ylabel("k-th moment", fontsize=15)
plt.legend(loc="best")
```

```python
plt.figure(figsize=(12, 9), dpi=80, facecolor="white")

plt.axhline(y=0, xmin=0, xmax=1, c="k", lw=1)

plt.plot(np.arange(1,5), diffs[0], label="Distr. 2 - Distr. 1", ls="", marker="^", markersize=10)
plt.plot(np.arange(1,5), diffs[1], label="Distr. 3 - Distr. 1", ls="", marker=">", markersize=10)
plt.plot(np.arange(1,5), diffs[2], label="Distr. 4 - Distr. 1", ls="", marker="v", markersize=10)
plt.plot(np.arange(1,5), diffs[3], label="Distr. 5 - Distr. 1", ls="", marker="<", markersize=10)

plt.xlabel("k", fontsize=20)
plt.ylabel("Difference in k-th moment", fontsize=20)
plt.xticks([1,2,3,4],fontsize=15)
plt.yticks(fontsize=15)
plt.legend(loc="best", fontsize=15, shadow=True)
```
